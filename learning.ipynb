```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='serif');
# Set seed for reproducibility
np.random.seed(42)

# Define the area dimensions
area_width = 100
area_height = 100

# Number of sensors
num_sensors = 200

# Generate sensor locations using Poisson Point Process
lambda_val = num_sensors / (area_width * area_height)
num_sensors_poisson = np.random.poisson(lambda_val * area_width * area_height)
x_poisson = np.random.uniform(0, area_width, num_sensors_poisson)
y_poisson = np.random.uniform(0, area_height, num_sensors_poisson)

# Generate sensor locations using Binomial Distribution
probability = num_sensors / (area_width * area_height)
num_sensors_binomial = np.random.binomial(area_width * area_height, probability)
x_binomial = np.random.uniform(0, area_width, num_sensors_binomial)
y_binomial = np.random.uniform(0, area_height, num_sensors_binomial)

# Plotting
plt.figure(figsize=(12, 12))

plt.subplot(2, 2, 1)
plt.scatter(x_poisson, y_poisson, marker='o', color='blue', alpha=0.5)
plt.title('Sensor Deployment - Poisson Point Process')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')

plt.subplot(2, 2, 2)
plt.scatter(x_binomial, y_binomial, marker='o', color='green', alpha=0.5)
plt.title('Sensor Deployment - Binomial Distribution')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')

num_sensors_poisson = np.random.poisson(lambda_val * area_width * area_height, size=10000)
num_sensors_binomial = np.random.binomial(area_width * area_height, probability, size=10000)


plt.subplot(2, 2, 3)
plt.hist(num_sensors_poisson, bins=np.arange(0, np.max(num_sensors_poisson) + 1) - 0.5, color='blue', edgecolor='black')
plt.title('Poisson Point Process')
plt.xlabel('Number of Sensors in a Cell')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.hist(num_sensors_binomial, bins=np.arange(0, np.max(num_sensors_binomial) + 1) - 0.5, color='green', edgecolor='black')
plt.title('Binomial Distribution')
plt.xlabel('Number of Sensors in a Cell')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
def poissonian(lambda_val, T, delta_n):
    sum = 0
    for n in range(delta_n + 1):
        sum += (((lambda_val * T)**n) * np.exp(-lambda_val * T))/np.math.factorial(n)

    return sum
```
